```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
model = nn.Linear(10, 2)
```

```python
batch = [torch.rand(32,10), torch.rand(32,2)]
```

```python
n_epochs = 1000
warm_up_steps = 50
lr = 3e-2
clip_grad = 1.0

# dummy
train_loader = [1]

T_max = len(train_loader) * n_epochs - warm_up_steps

optimizer = optim.SGD(model.parameters(), lr=0.03, momentum=0.9)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max, eta_min=0, last_epoch=-1, verbose=False)
criterion = nn.MSELoss()
```

```python
for i in range(n_epochs):
    
    optimizer.zero_grad()
    
    if i < warm_up_steps:
        learning_rate = lr * (i+1)/warm_up_steps
        optimizer.param_groups[0]["lr"] = learning_rate
    
    out = model(batch[0])
    
    loss = criterion(out, batch[1])
    print(f'loss {loss.item()} lr {optimizer.param_groups[0]["lr"]}')
    loss.backward()

    if clip_grad != 0:
        nn.utils.clip_grad_norm_(model.parameters(), clip_grad)
    
    optimizer.step()

    if i >= warm_up_steps:
        scheduler.step()
```
